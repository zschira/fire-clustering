# Fire Clustering

```python
import numpy as np
import scipy.cluster.hierarchy as hcluster
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
```

## Import Data

```python
rIN = open(r"days4.txt",'r')
dayText = rIN.readlines()
rIN.close()
dayList = [l.split('\t') for l in dayText]
days = np.array(dayList,dtype=int)
```

## Find pixels where fire occured

```python
fires = np.where(days > -1)
fires = np.vstack((fires, days[days > -1]))
plt.scatter(fires[0,:], fires[1,:])
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.scatter(fires[0,:], fires[1,:], fires[2,:])
```

## Find Clusters

```python
thresh = 2.5
clusters = hcluster.fclusterdata(np.transpose(fires), thresh, criterion="distance")
```

## Plot with clusters labeled

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.scatter(fires[0,:], fires[1,:], fires[2,:], c=clusters)
plt.show()
```
